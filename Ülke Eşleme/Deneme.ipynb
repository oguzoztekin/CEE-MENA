```python
import pandas as pd

df_2= pd.read_csv(r"country_codes_V202401.csv")
df_1 = pd.read_excel(r"Ülkeler ve iso_3 Kodları.xlsx")
```

```python
import pandas as pd


# İlk dosyanın 'country_name' sütunu oluşturuluyor
df_1['country_name'] = df_1['Real GDP growth (Annual percent change)']

# Birleştirme işlemi: İlk dosyadaki tüm verileri koruyarak
merged_df = pd.merge(df_1, df_2, left_on='country_name', right_on='country_name', how='left')

# Gerekli sütunları seçme
result = merged_df[['country_code', 'country_iso3', 'Real GDP growth (Annual percent change)']]

# Sonuçları kaydetmek veya göstermek için
result.to_excel("merged_output.xlsx", index=False)  # Sonucu bir Excel dosyasına kaydedebilirsiniz
print(result)
```
